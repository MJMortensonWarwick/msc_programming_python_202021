```python
import pandas as pd
import numpy as np

jan_df = pd.read_excel('fake_sales.xlsx', sheetname='January')
```

```python
jan_df.head()
```

```python
print(jan_df.dtypes)
```

```python
feb_df = pd.read_excel('fake_sales.xlsx', sheetname='February')
```

```python
feb_df.head()
```

```python
feb_df.columns = ['Date', 'US_Sales', 'UK_Sales', 'Canada_Sales']
feb_df.head()
```

```python
sales_df = pd.concat([jan_df, feb_df], axis=0, join='outer')
```

```python
sales_df
```

```python
sales_df = sales_df.fillna(0)
sales_df
```

```python
sales_df['US_Sales'] = sales_df['US_Sales'] * 0.75
sales_df['Canada_Sales'] = sales_df['Canada_Sales'] * 0.56
sales_df
```

```python
sales_df['Month'] = pd.DatetimeIndex(sales_df['Date']).month
```

```python
sales_df
```

```python
sales_df['Month'] = np.where(sales_df.Month == 1, 'Jan', 'Feb')
```

```python
sales_df
```

```python
sales_group_df = sales_df.groupby(['Month']).mean()
```
