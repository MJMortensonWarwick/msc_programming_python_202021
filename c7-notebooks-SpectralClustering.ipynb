```python
# import the generator and create the data
from sklearn.datasets import make_circles
X, y = make_circles(n_samples=100, factor=.3) # factor determines the gap between circles 

# visualise the data
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1], color = 'g') 
plt.show() 
```

```python
# import and fit the clustering algorithm
from sklearn.cluster import SpectralClustering
clustering = SpectralClustering(n_clusters=2, affinity="nearest_neighbors").fit(X)
```

```python
# create a Dataframe using the values of X and the cluster labels generated
import pandas as pd
df = pd.DataFrame(data={'x1': X[:, 0], 'x2': X[:, 1], 'label': clustering.labels_})  

# visualise the cluster analysis with Seaborn
import seaborn as sns
sns.scatterplot(data=df, x='x1', y='x2', hue=df.label.to_list())
```
